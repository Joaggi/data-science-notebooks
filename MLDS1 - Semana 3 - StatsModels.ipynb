# Regresión lineal con variables dummy

```python
import pandas as pd
```

```python
nsample = 50
groups   = np.zeros(nsample, int)
groups[20:40] = 1
groups[40:] = 2
groups
```

```python
df = pd.get_dummies(groups)
display(df)
dummy = pd.get_dummies(groups).values
dummy
```

```python
x = np.linspace (0, 20, nsample)
x
```

```python
X = np.column_stack((x, dummy[:, 1:]))
X
```

```python
import statsmodels.api as sm
X = sm.add_constant(X, prepend=False)
X
```

```python
beta = [1., 3, -3, 10]
y_true = np.dot(X, beta)
e = np.random.normal(size=nsample)
y = y_true + e
```

```python
res2 = sm.OLS(y, X).fit()
res2.summary()
```

```python
import numpy as np
def elevar(dato):
  return np.log(dato) * np.exp(dato)

def funcion1():
  lista = np.arange(100)

  for i in lista:
    elevar(i)

%timeit funcion1()
```

# Explicación de filtros compuestos

```python
import pandas as pd

df = pd.DataFrame({'País': ["Colombia", "USA", "Chile", "Argentina"],
                   'Ciudad': ["Bogotá", "Miami", "Valpaiso", "Buenos Aires"],
                   'Población': [10000, 20000, 3000, 1000]})

df
```

La forma menos recomendable de acceder a un elemento de un dataframe es con el punto, ya que cabe la posibilidad de que el nombre de la columna empiece con un caracter no valido para el nombramiento de variables de Python

```python
df.Población
```

La forma más recomendable es acceder como si fuese un diccionario.

```python
df["Población"]
```

Yo puedo definir una variable y en ella guardar el nombre de la columna que estoy buscando

```python
variable = "Población"
df[variable]
```

Si yo quiero filtrar por un indice especifico y una columna lo puedo hacer con la función loc

```python
df.loc[1, "Población"]
```

```python
variable = "Población"
df.loc[1, variable]
```

```python
def buscar_por_columna(data_frame, columna):
  return data_frame[columna]

buscar_por_columna(df, "Población")
```

```python
def buscar_por_columna(data_frame, columna):
  # con los : le puedo pedir a pandas que devuelva todos los elementos dada una columna
  return data_frame.loc[:, columna]

buscar_por_columna(df, "Población")
```

Como buscar un elemento especifico

Cuando aplico una operación de tipo relacional (<,>,<=,>=,==,!=) en un dataframe de Pandas, el me devuelve una Serie de booleanos, con tantas filas como filas tiene el dataframe original

```python
df["Ciudad"] == "Miami"
```

```python
filtro =  df["Ciudad"] == "Miami"
```

Yo puedo utilizar el filtro directamente en pandas con [], sin embargo es preferible hacerlo con la función .loc ya que hace más fácil su entendimiento.

```python
df[filtro]
```

Recuerden que el primer elemento de .loc son las filas y el segundo separado por coma son las columnas.

```python
df.loc[filtro]
```

Cuando necesitamos hacer un filtro compuesto Pandas tomo la decisión de que para realizarlo necesitamos cambiar el or por (|, pleca), and por (&, ampersand), not por (~, virgulilla).

Por ejemplo, si yo quiero filtrar por población que esta entre 3 mil y 10 mil lo haríamos como:

Se podría llegar a pensar que la operacíon funcionaría de la siguiente manera.

```python
df.loc[df["Población"]>=3000 & df["Población"]<=10000 ] # se genera error
```

Sin embargo, Pandas exige que cada una de los filtros se enmarque dentro de paréntesis ().

```python
df.loc[(df["Población"]>=3000) & (df["Población"]<=10000 )]
```

```python
df.loc[(df["Población"]<3000) | (df["Población"]>10000 )]
```

```python
df.loc[((df["Población"]<3000) | (df["Población"]>10000 )) & (df["Ciudad"]=="Miami" )]
```

¿Cuando tenemos un máximo o un minimo que pasa?

Cuando se aplica la función minimo, se obtiene un solo valor sobre la columna

```python
df["Población"].min()
```

```python
df[df["Población"].min()] # Esto genera error ya que no se esta pasando ni un indice ni una serie
```

```python
df["Población"] == df["Población"].min()
```

```python
df[df["Población"] == df["Población"].min()]
```

```python
df.loc[df["Población"] == df["Población"].min()]
```

Como eliminar luego de encontrar un elemento especifico

Con el filtro de eliminación yo puedo filtrar para obtener los indices en los que cierto filtro es verdadero

```python
filtro_eliminacion = df["Población"] == df["Población"].min()
df[filtro_eliminacion].index
```

```python
df.drop(df[filtro_eliminacion].index)
```

Algo que se le puede venir a la mente es utilizar la función argmin y utilizarlo directamente en drop, sin embargo esto solo eliminaría la primera ocurrencia que es encuentre del valor minimo.

```python
df["Población"].argmin()
```

# Ejemplo función Mask

```python
import numpy as np
import pandas as pd

data = pd.DataFrame(np.random.normal(size=(100,2)), columns=["x", "y"])
data.loc[0:9, "y"] = 10

data
```

```python
mean_y = data["y"].mean()
std_y = data["y"].std()
print(f"mean_y: {mean_y} std_y: {std_y}")
```
